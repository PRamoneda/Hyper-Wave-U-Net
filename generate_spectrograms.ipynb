```python
# import libraries
import librosa
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
```

```python
# paths to files: true source, mhe and baseline estimations
true_path = r'C:\Users\Joaquin\Documents\GitHub\Wave-U-Net\Source_Estimates\examples\Motor Tapes - Shore'
mhe_path = r'C:\Users\Joaquin\Documents\GitHub\Wave-U-Net\Source_Estimates\examples\Motor Tapes - Shore\Motor Tapes - Shore - mhe_0'
base_path = r'C:\Users\Joaquin\Documents\GitHub\Wave-U-Net\Source_Estimates\examples\Motor Tapes - Shore\Motor Tapes - Shore - baseline'

# filenames: true source, mhe and baseline estimations
true = 'Motor Tapes - Shore.stem_vocals.wav'
mhe = 'vocals.wav'
base = 'vocals.wav'
```

```python
# Print True Spectrogram
x, sr = librosa.load(true_path + '/' + true)
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
plt.title('true vocals')
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
```

```python
# Print MHE Spectrogram
x_mhe, sr_mhe = librosa.load(mhe_path + '/' + mhe)
X_mhe = librosa.stft(x_mhe)
Xdb_mhe = librosa.amplitude_to_db(abs(X_mhe))
plt.figure(figsize=(14, 5))
plt.title('mhe_0 model estimated vocals')
librosa.display.specshow(Xdb_mhe, sr=sr_mhe, x_axis='time', y_axis='hz')
```

```python
# Print base Spectrogram
x_base, sr_base = librosa.load(base_path + '/' + base)
X_base = librosa.stft(x_base)
Xdb_base = librosa.amplitude_to_db(abs(X_base))
plt.figure(figsize=(14, 5))
plt.title('baseline model estimated vocals')
librosa.display.specshow(Xdb_base, sr=sr_base, x_axis='time', y_axis='hz')
```
